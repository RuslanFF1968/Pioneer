```
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(book)
print(tokens)
```

```
tokenizer = RegexpTokenizer(r'\w+')
tokens = [token.lower() for token in tokenizer.tokenize(book)]
print(tokens)
```

```
words_parsed = [(morph.parse(word), word) for word in tokens]
```

```
words=[]
for word in words_parsed:
  words.append(
      {
          "lemma": word[0][0].normal_form,
          "word": word[1],
          "pos": word[0][0].tag.POS,
      }
  )
words
```

```
data = pd.DataFrame(words)
data.head()
```

```
q=data['pos'].dropna().value_counts(normalize=True)*100
q.head(10).plot(kind="bar")
```

```
infinitives=data[data['pos']=='VERB']['lemma']

wordcloud = WordCloud(width=400, height=400, margin=0).generate(" ".join(infinitives))
infinitives.value_counts().head(30)
```

```
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0)
plt.show()
```

```
bigramms = list(nltk.bigrams(data['lemma']))
sr = pd.Series(" ".join(a) for a in bigramms)
sr.value_counts().head(25)
bigramms
```

```
trigramms = list(nltk.trigrams(data['lemma']))
sr = pd.Series(" ".join(a) for a in trigramms)
sr.value_counts().head(25)
```

```
def change_time(s):
  try:
    word=morph.parse(s)[0]
    if word.tag.POS=="WERB":
      s = word.inflect({'pres' if word.tag.tense == 'past' else 'past'}).word
  except (AttributeError, KeyError, ValueError):
    pass
  return s

def change_plural(s):
  try:
    word = morph.parse(s)[0]
    s = word.inflect({'plur' if word.tag.number == 'sing' else 'sing'}).word
  except (AttributeError, KeyError, ValueError):
    pass
  return s

s="""Без всякого шума и суматохи она обошла весь дом, добралась наконец до лестницы и, никому ничего не сказав, одна поднялась на три пролета, вошла в свою комнату и молча легла, как старинная мумия"""
s_split=s.split()
for word in range(len(s_split)):
 s_split[word] = change_time(s_split[word])
 s_split[word] = change_plural(s_split[word])

" ".join(s_split)
```
