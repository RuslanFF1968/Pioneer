## 1. Используя код с занятия создайте key-value storage на 1 миллион ключей.

````
# Создадим Key-Value (HashTable) storage на 1 млн объектов
# Ключем будет являтс id объекта от 0 до 1.000.000, значением будет являтся пустой массив
kv_storage = {i:[] for i in range(1000000)}

print('Key-Value storage =', len(kv_storage))
print('Проверить первый элемент  =', kv_storage[0])
````

## 2. Заполните его случайными значениями

````
# Заполним наше ПУСТОЕ хранилище для каждого ключа произвольным количеством событий со случайными значениями:
import numpy as np

for k in kv_storage.keys():
  num_events = np.random.randint(0,320) # создаем случайное число
  new_val = np.random.uniform(low=0,high=50,size=num_events).tolist() # создаем массив размером в случайное число и преобразуем в список из numpy
  kv_storage[k].extend(new_val)
````

## 3. Придумайте 3 любые метрики по аналогии с метриками с урока

````
# Придумайте 3 любые метрики
# метрики - стандартное отклонение, занчения больше 100, минимум и максимум
# допишите в класс KpiBuilder функции
#Теперь напишем простенький класс для расчета KPI
class KpiBuilder():
  def __init__(self, key):
        self.key = key

  def get(self):
    return kv_storage.get(self.key)

# Среднее значение метрик за последние 20 отсчетов
  def mean_20(self):
    return sum(self.get()[:-20])/20.

# Медианное значение метрик за весь период
  def median(self):
    m_ = self.get()
    return np.median(m_)

# Количество не нулевых метрик
  def non_zero_cnt(self):
    v = self.get()
    return len([x for x in v if x > 0])

# Количество метрик превышающих среднее по всем элементам
  def cnt_above_mean(self):
    v = self.get()
    m_ = kv_storage['sum_el']/ float(kv_storage['len_el'])

    return len([x for x in v if x > m_])

# Среднеквадратическое отклонение
  def std(self):
    return np.std(kv_storage[self.key])

# Кол-во метрик в диапазоне 10% от среднеквадратического отклонения
  def cnt_above_std(self):
    v = self.get()
    a = self.std()
    return len([x for x in v if (x < a*1.1) and (x > a*0.9)])

# Разница между max & min из значений, посчитанных в предыдущей метрике
  def MaxMin(self):
    v = self.get()
    a = self.std()
    list = [x for x in v if (x < a*1.1) and (x > a*0.9)]
    return max(list)-min(list)
````

````
# Инициализируем вспомогательное значение sum_el/ len_el суммой элементов
v_len = len([item for sublist in kv_storage.values() for item in sublist])
v_sum = sum([item for sublist in kv_storage.values() for item in sublist])

kv_storage['len_el']=v_len
kv_storage['sum_el']=v_sum

# для сравнения
print('Длина:', kv_storage['len_el'])
print('Сумма:', kv_storage['sum_el'])
````

##Продемонстрируйте расчеты

````
key_120 = KpiBuilder(120)

print(f"Среднее значение метрик за последние 20 отсчетов = {key_120.mean_20()}")
print(f"Медианное значение метрик за весь период = {key_120.median()}")
print(f"Количество не нулевых метрик = {key_120.non_zero_cnt()}")
print(f"Количество метрик превышающих среднее по всем элементам = {key_120.cnt_above_mean()} ", '\n')

print(f"Среднеквадратическое отклонение   = {key_120.std()}")
print(f"Кол-во метрик в диапазоне 10 процентов от среднеквадратического отклонения = {key_120.cnt_above_std()}")
print(f"Разница между max & min = {key_120.MaxMin()}")
````
